##### 1. Write a program to visualize the total profit of all months using line plot. 2. Write a program to visualize total profit of all months and show line plot with the following Style properties. 
-> Line Style dotted and Line-color should be blue
-> Show legend at the upper right location.
-> X label name = Month Number
-> Y label name = Item Sold units number
-> Add a circle marker
-> Line marker color as yellow
-> Line width should be 5 

```python
plt.plot(df.month_name,df.total_profit,marker='o',color='b',linestyle='--',linewidth=5,markerfacecolor='y',label='Total_Profit_per_Month')
plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.legend(loc='upper right')
```
